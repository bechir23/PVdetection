```python
import os
def remove_segmentations(image_dir,label_dir):
    c=0
    for label_file in os.listdir(label_dir):

        if label_file.endswith(".txt"):
            label_path = os.path.join(label_dir, label_file)
            with open(label_path, 'r') as file:
                lines = file.readlines()
            if any(len(line.split()) > 5 for line in lines):
                print(f"Removing segment from {label_file}")
                os.remove(label_path)
                c=c+1

                # Assuming image has the same name but different extension
                # Adjust the extension based on your dataset (e.g., .jpg, .png)
                image_file = label_file.replace(".txt", ".jpg")
                image_path = os.path.join(image_dir, image_file)

                if os.path.exists(image_path):
                    os.remove(image_path)  # Remove corresponding image file
                    print(f"Removed image file {image_file}")
                else:
                    print(f"Image file {image_file} not found.")
    return c


image_dir = 'Photovoltaic-1/valid/images'
label_dir = 'Photovoltaic-1/valid/labels'
print(remove_segmentations(image_dir,label_dir))
image_dir = 'Photovoltaic-1/train/images'
label_dir = 'Photovoltaic-1/train/labels'
print(remove_segmentations(image_dir,label_dir))
image_dir = 'Photovoltaic-1/test/images'
label_dir = 'Photovoltaic-1/test/labels'
print(remove_segmentations(image_dir,label_dir))
```

```python
import os
def predict_and_append(model,image_dir,label_dir):
  
  for image_file in os.listdir(image_dir):
    if image_file.endswith(".jpg"):
      image_path=os.path.join(image_dir,image_file)
      results=model.predict(image_path)
      results_json=results.json()
      items=[item for item in results_json['predictions'] ]
      if items:
        label_file=image_file.replace(".jpg",".txt")
        label_path=os.path.join(label_dir,label_file)
        with open(label_path,'a') as f:

          for item in items:
           if item['class']=='3':

            center_x = item['x']/640
            center_y = item['y']/640
            width = item["width"]/640
            height = item["height"]/640
            class_id = 1

            line = f"\n{class_id} {center_x} {center_y} {width} {height}"
            f.write(line)
            print(f"Predictions appended in YOLO format for {label_path}")


image_dir = 'Photovoltaic-1/valid/images'
label_dir = 'Photovoltaic-1/valid/labels'
predict_and_append(model,image_dir,label_dir)
image_dir = 'Photovoltaic-1/train/images'
label_dir = 'Photovoltaic-1/train/labels'
predict_and_append(model,image_dir,label_dir)

image_dir = 'Photovoltaic-1/test/images'
label_dir = 'Photovoltaic-1/test/labels'
predict_and_append(model,image_dir,label_dir)
```
